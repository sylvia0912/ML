## Objective Functions/Loss function/Cost Function in ML

Reference - https://medium.com/@bhanuyerra/objective-functions-used-in-machine-learning-9653a75363b5

http://kronosapiens.github.io/blog/2017/03/28/objective-functions-in-machine-learning.html

- Most of ML is about Optimization problems, as the name implies, deal with finding the best, or “optimal” (hence the name) solution to some type of problem, generally mathematical.
- In order to find the optimal solution, we need some way of measuring the quality of any solution. This is done via what is known as an **objective function, with “objective” used in the sense of a goal.**
- This function, taking data and model parameters as arguments, can be evaluated to return a number. Any given problem contains some parameters which can be changed; our goal is to find values for these parameters which either maximize or minimize this number.
- **For some objectives, the optimal parameters can be found exactly (known as the analytic solution). For others, the optimal parameters cannot be found exactly (known as numerical solutions), but can be approximated using a variety of iterative algorithms**

- This means that optimization methods used are different.E.g OLS (Ordinary least squares) is an optimization method (since it also works on MINIMIZATION of residual errors) that gets us an analytical solution (exact solution) for linear regression problem while u can also use other optimization methods like gradient descent(iterative algorithms) which are used for computationally harder problems where instead of exact solution we get to best approximate solution. It is not that optimatization methods like gradient descent can only be applied to non-linear or complex problems u can also apply this optimization method to linear regression & it should converge to same solution as OLS, but faster.
- If u run OLS optimization (analytic solution) multiple times, it should give same result. If u run gradient descent optimization (numerical solution) multiple times it should give different results each time.

**Note**

Further when using numerical solution, if non-convex problem (has multiple local minima) below methods can be used. Generally in convex problems the global and local minima are same.
